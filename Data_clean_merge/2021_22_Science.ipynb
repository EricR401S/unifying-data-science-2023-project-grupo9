```python
import pandas as pd
import numpy as np

df_annual_em_science_2022 = pd.read_csv(
    "/Users/kashafali/Documents/Duke/Spring23/UDS/Project/Data/2022/Annual EM Science.csv"
)
```

```python
df_annual_em_science_2022.columns
```

# Dataframe for All Students

```python
# only keep columns
column_names = [
    "ENTITY_CD",
    "ENTITY_NAME",
    "YEAR",
    "ASSESSMENT_NAME",
    "SUBGROUP_NAME",
    "TOTAL_COUNT",
    "NOT_TESTED",
    "PCT_NOT_TESTED",
    "NUM_TESTED",
    "PCT_TESTED",
    "NUM_PROF",
    "PER_PROF",
    "TOTAL_SCALE_SCORES",
    "MEAN_SCORE",
]

df_annual_em_science_2022_total = df_annual_em_science_2022[column_names]
df_annual_em_science_2022_total = df_annual_em_science_2022_total[
    df_annual_em_science_2022_total["SUBGROUP_NAME"] == "All Students"
]
```

```python
df_annual_em_science_2022_total.head()
```

```python
df_annual_em_science_2022_merge = df_annual_em_science_2022_total.drop(
    columns=["SUBGROUP_NAME"]
).copy()
```

# Dataframe for Gender

```python
df_annual_em_science_2022_gender = df_annual_em_science_2022.copy()

df_annual_em_science_2022_gender = df_annual_em_science_2022_gender[column_names]

df_annual_em_science_2022_gender = df_annual_em_science_2022_gender[
    (df_annual_em_science_2022_gender["SUBGROUP_NAME"] == "Female")
    | (df_annual_em_science_2022_gender["SUBGROUP_NAME"] == "Male")
]
```

```python
# fix index
df_annual_em_science_2022_gender = df_annual_em_science_2022_gender.set_index(
    ["SUBGROUP_NAME", "ENTITY_CD", "YEAR", "ASSESSMENT_NAME"]
).unstack(["SUBGROUP_NAME"])

df_annual_em_science_2022_gender = df_annual_em_science_2022_gender.reindex(
    columns=sorted(df_annual_em_science_2022_gender.columns, key=lambda x: x[::-1])
)

df_annual_em_science_2022_gender.columns = [
    "{}_{}".format(t, v) for v, t in df_annual_em_science_2022_gender.columns
]

df_annual_em_science_2022_gender = pd.DataFrame(df_annual_em_science_2022_gender.to_records())
```

```python
df_annual_em_science_2022_gender.head()
```

# Merge Gender Dataframe with All Students Dataframe

```python
df_science_2022_1 = pd.merge(
    df_annual_em_science_2022_merge,
    df_annual_em_science_2022_gender,
    on=["ENTITY_CD", "YEAR", "ASSESSMENT_NAME"],
    how="left",
)

df_science_2022_1.head()
```

# Dataframe for Race

```python
df_annual_em_science_2022_race = df_annual_em_science_2022[column_names].copy()

df_annual_em_science_2022_race = df_annual_em_science_2022_race[
    (df_annual_em_science_2022_race["SUBGROUP_NAME"] == "American Indian or Alaska Native")
    | (df_annual_em_science_2022_race["SUBGROUP_NAME"] == "Black or African American")
    | (df_annual_em_science_2022_race["SUBGROUP_NAME"] == "Hispanic or Latino")
    | (
        df_annual_em_science_2022_race["SUBGROUP_NAME"]
        == "Asian or Native Hawaiian/Other Pacific Islander"
    )
    | (df_annual_em_science_2022_race["SUBGROUP_NAME"] == "White")
    | (df_annual_em_science_2022_race["SUBGROUP_NAME"] == "Multiracial")
]
```

```python
# reindex

df_annual_em_science_2022_race = df_annual_em_science_2022_race.set_index(
    ["SUBGROUP_NAME", "ENTITY_CD", "YEAR", "ASSESSMENT_NAME"]
).unstack(["SUBGROUP_NAME"])

df_annual_em_science_2022_race = df_annual_em_science_2022_race.reindex(
    columns=sorted(df_annual_em_science_2022_race.columns, key=lambda x: x[::-1])
)

df_annual_em_science_2022_race.columns = [
    "{}_{}".format(t, v) for v, t in df_annual_em_science_2022_race.columns
]

df_annual_em_science_2022_race = pd.DataFrame(df_annual_em_science_2022_race.to_records())
```

```python
df_annual_em_science_2022_race.head()
```

# Merge Race Dataset with All Students + Gender Dataset

```python
df_science_2022_2 = pd.merge(
    df_science_2022_1,
    df_annual_em_science_2022_race,
    on=["ENTITY_CD", "YEAR", "ASSESSMENT_NAME"],
    how="left",
)

df_science_2022_2.head()
```

# English Language Learners Dataframe

```python
df_annual_em_science_2022_eng = df_annual_em_science_2022[column_names].copy()


df_annual_em_science_2022_eng = df_annual_em_science_2022_eng[
    (df_annual_em_science_2022_eng["SUBGROUP_NAME"] == "Non-English Language Learner")
    | (df_annual_em_science_2022_eng["SUBGROUP_NAME"] == "English Language Learner")
]
```

```python
# reindex

df_annual_em_science_2022_eng = df_annual_em_science_2022_eng.set_index(
    ["SUBGROUP_NAME", "ENTITY_CD", "YEAR", "ASSESSMENT_NAME"]
).unstack(["SUBGROUP_NAME"])

df_annual_em_science_2022_eng = df_annual_em_science_2022_eng.reindex(
    columns=sorted(df_annual_em_science_2022_eng.columns, key=lambda x: x[::-1])
)

df_annual_em_science_2022_eng.columns = [
    "{}_{}".format(t, v) for v, t in df_annual_em_science_2022_eng.columns
]

df_annual_em_science_2022_eng = pd.DataFrame(df_annual_em_science_2022_eng.to_records())
```

```python
df_annual_em_science_2022_eng.head()
```

# Merge English Language Learners Dataset with All Students + Gender + Race Dataset

```python
df_science_2022_3 = pd.merge(
    df_science_2022_2,
    df_annual_em_science_2022_eng,
    on=["ENTITY_CD", "YEAR", "ASSESSMENT_NAME"],
    how="left",
)

df_science_2022_3.head()
```

# Dataframe for Economic Condition

```python
df_annual_em_science_2022_econ = df_annual_em_science_2022[column_names].copy()

df_annual_em_science_2022_econ = df_annual_em_science_2022_econ[
    (df_annual_em_science_2022_econ["SUBGROUP_NAME"] == "Economically Disadvantaged")
    | (df_annual_em_science_2022_econ["SUBGROUP_NAME"] == "Not Economically Disadvantaged")
]
```

```python
# reindex

df_annual_em_science_2022_econ = df_annual_em_science_2022_econ.set_index(
    ["SUBGROUP_NAME", "ENTITY_CD", "YEAR", "ASSESSMENT_NAME"]
).unstack(["SUBGROUP_NAME"])

df_annual_em_science_2022_econ = df_annual_em_science_2022_econ.reindex(
    columns=sorted(df_annual_em_science_2022_econ.columns, key=lambda x: x[::-1])
)

df_annual_em_science_2022_econ.columns = [
    "{}_{}".format(t, v) for v, t in df_annual_em_science_2022_econ.columns
]

df_annual_em_science_2022_econ = pd.DataFrame(df_annual_em_science_2022_econ.to_records())
```

```python
df_annual_em_science_2022_econ.head()
```

# Merge Economic Dataset with All Students + Gender + Race + English Language Dataset

```python
df_science_2022_4 = pd.merge(
    df_science_2022_3,
    df_annual_em_science_2022_econ,
    on=["ENTITY_CD", "YEAR", "ASSESSMENT_NAME"],
    how="left",
)

df_science_2022_4.head()
```

# Dataframe for Special Education

```python
df_annual_em_science_2022_dis = df_annual_em_science_2022[column_names].copy()

df_annual_em_science_2022_dis = df_annual_em_science_2022_dis[
    (df_annual_em_science_2022_dis["SUBGROUP_NAME"] == "Students with Disabilities")
]
```

```python
# reindex

df_annual_em_science_2022_dis = df_annual_em_science_2022_dis.set_index(
    ["SUBGROUP_NAME", "ENTITY_CD", "YEAR", "ASSESSMENT_NAME"]
).unstack(["SUBGROUP_NAME"])

df_annual_em_science_2022_dis = df_annual_em_science_2022_dis.reindex(
    columns=sorted(df_annual_em_science_2022_dis.columns, key=lambda x: x[::-1])
)

df_annual_em_science_2022_dis.columns = [
    "{}_{}".format(t, v) for v, t in df_annual_em_science_2022_dis.columns
]

df_annual_em_science_2022_dis = pd.DataFrame(df_annual_em_science_2022_dis.to_records())
```

```python
df_annual_em_science_2022_dis.head()
```

# Merge Special Education Dataset with All Students + Gender + Race + English Language + Econ Dataset

```python
df_science_2022_5 = pd.merge(
    df_science_2022_4,
    df_annual_em_science_2022_dis,
    on=["ENTITY_CD", "YEAR", "ASSESSMENT_NAME"],
    how="left",
)

df_science_2022_5.head()
```
